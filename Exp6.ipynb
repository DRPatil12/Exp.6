```python
#9. Salary Calculation with allowances:
'''Problem statement: Write a python program to calculate the gross salary
of an employee.
If basic =< 10000: HRA = 20%, DA = 80%
If basic =< 20000: HRA = 25%, DA = 90%
otherwise: HRA = 30%, DA = 95%'''

basic = float(input("Enter Salary:"))
if basic<= 10000:
  HRA = basic*0.20
  DA = basic*0.80
elif basic <= 20000:
  HRA = basic*0.25
  DA = basic*0.90
else:
  HRA = basic*0.30
  DA = basic*0.95
gross = basic + HRA + DA
print("Gross Salary:", gross)
```

```python
'''10. Write a python program to calculate income tax based on annual income:
Up to rupees 2,50,000: No tax
2,50,001 - 5,00,000: 5%
5,00,001 - 10,00,000: 20%
above 10,00,000: 30%'''
income = float(input("Enter annual income:"))
if income <= 250000:
  tax = 0
elif income <= 500000:
  tax = (income - 250000)*0.05
elif income <= 1000000:
  tax = (250000*0.05) + ( income - 500000)*0.20
else:
  tax = (250000*0.05) + (500000*0.20) + (income - 1000000)*0.30

print("Income Tax:", tax)
```
